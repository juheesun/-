````python
import pandas as pd

# ì„±ìˆ˜ê¸° êµ¬ë¶„ ë°ì´í„° ìƒì„± í•¨ìˆ˜
def generate_peak_season_data(df):
    # ì—…ì¥+ë©”ë‰´ ë‹¨ìœ„ë¡œ ì›”ë³„ í‰ê·  ë§¤ì¶œ ê³„ì‚°
    monthly_avg = (
        df.groupby(["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…", "ì›”"])["ë§¤ì¶œìˆ˜ëŸ‰"]
        .mean()
        .reset_index()
    )

    # ê° ì¡°í•©ë³„ í‰ê·  ë§¤ì¶œ
    base_avg = (
        monthly_avg.groupby(["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…"])["ë§¤ì¶œìˆ˜ëŸ‰"]
        .mean()
        .reset_index()
        .rename(columns={"ë§¤ì¶œìˆ˜ëŸ‰": "ê¸°ì¤€ë§¤ì¶œ"})
    )

    # ë³‘í•© í›„ ì„±ìˆ˜ê¸°/ë¹„ìˆ˜ê¸° êµ¬ë¶„ (ì„ê³„ì¹˜: í‰ê·  ëŒ€ë¹„ 1.3ë°° ì´ìƒ = ì„±ìˆ˜ê¸°, 0.7ë°° ì´í•˜ = ë¹„ìˆ˜ê¸°)
    monthly_labeled = pd.merge(monthly_avg, base_avg, on=["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…"])
    monthly_labeled["ì„±ìˆ˜ê¸°ì—¬ë¶€"] = monthly_labeled.apply(
        lambda row: "ì„±ìˆ˜ê¸°" if row["ë§¤ì¶œìˆ˜ëŸ‰"] >= row["ê¸°ì¤€ë§¤ì¶œ"] * 1.3
        else "ë¹„ìˆ˜ê¸°" if row["ë§¤ì¶œìˆ˜ëŸ‰"] <= row["ê¸°ì¤€ë§¤ì¶œ"] * 0.7
        else "êµ¬ë¶„ ë¶ˆê°€",
        axis=1
    )
    # ê²°ê³¼: ì—…ì¥+ë©”ë‰´+ì›”ë³„ ì„±ìˆ˜ê¸° ì •ë³´
    return monthly_labeled[["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…", "ì›”", "ì„±ìˆ˜ê¸°ì—¬ë¶€"]].copy()


def preprocess_sales_data_first(df):
    # ë‚ ì§œ ë³€í™˜
    df["ì˜ì—…ì¼ì"] = pd.to_datetime(df["ì˜ì—…ì¼ì"])

    # ì˜ì—…ì¥ëª…ê³¼ ë©”ë‰´ëª… ë¶„ë¦¬
    df["ì˜ì—…ì¥ëª…"] = df["ì˜ì—…ì¥ëª…_ë©”ë‰´ëª…"].apply(lambda x: x.split("_")[0])
    df["ë©”ë‰´ëª…"] = df["ì˜ì—…ì¥ëª…_ë©”ë‰´ëª…"].apply(lambda x: x.split("_")[1])

    # ìš”ì¼ ë° ì£¼ë§ ì—¬ë¶€
    df["ì›”"] = df["ì˜ì—…ì¼ì"].dt.month
    df["ìš”ì¼"] = df["ì˜ì—…ì¼ì"].dt.weekday
    df["ì£¼ë§ì—¬ë¶€"] = df["ìš”ì¼"].isin([5, 6]).astype(int)

    return df


def preprocess_sales_data_second(df, peak_season_map=None):
    # ê³µíœ´ì¼ í•˜ë“œì½”ë”© (2023~2025)
    holidays_2023_2025 = pd.to_datetime([
        # 2023ë…„
        "2023-01-01", "2023-01-21", "2023-01-22", "2023-01-23", "2023-03-01", "2023-05-05",
        "2023-05-27", "2023-06-06", "2023-08-15", "2023-09-28", "2023-09-29", "2023-09-30",
        "2023-10-03", "2023-10-09", "2023-12-25",
        # 2024ë…„
        "2024-01-01", "2024-02-09", "2024-02-10", "2024-02-11", "2024-03-01", "2024-05-05",
        "2024-05-06", "2024-06-06", "2024-08-15", "2024-09-16", "2024-09-17", "2024-09-18",
        "2024-10-03", "2024-10-09", "2024-12-25",
        # 2025ë…„
        "2025-01-01", "2025-01-28", "2025-01-29", "2025-01-30", "2025-03-01", "2025-05-05",
        "2025-06-06", "2025-08-15", "2025-10-03", "2025-10-06", "2025-10-07", "2025-10-08",
        "2025-10-09", "2025-12-25"
    ])
    df["ê³µíœ´ì¼ì—¬ë¶€"] = df["ì˜ì—…ì¼ì"].isin(holidays_2023_2025).astype(int)

    # ì„±ìˆ˜ê¸° ì—¬ë¶€ íŒë³„
    # ì›” ì»¬ëŸ¼ ìƒì„± í›„ ì„±ìˆ˜ê¸° ë§µê³¼ ë³‘í•©
    df["ì›”"] = df["ì˜ì—…ì¼ì"].dt.month
    df = df.merge(
        peak_season_map,
        how="left",
        on=["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…", "ì›”"]
    )

    # ì„±ìˆ˜ê¸°ì—¬ë¶€ ê²°ì¸¡ê°’ì€ "êµ¬ë¶„ ë¶ˆê°€" ì²˜ë¦¬
    df["ì„±ìˆ˜ê¸°ì—¬ë¶€"] = df["ì„±ìˆ˜ê¸°ì—¬ë¶€"].fillna("êµ¬ë¶„ ë¶ˆê°€")

    # time_idx ì¶”ê°€ (ì‹œê³„ì—´ ì¸ë±ìŠ¤)
    df = df.sort_values(["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…", "ì˜ì—…ì¼ì"])
    df["time_idx"] = df.groupby(["ì˜ì—…ì¥ëª…", "ë©”ë‰´ëª…"]).cumcount()

    # group_id ì¶”ê°€
    df["group_id"] = df["ì˜ì—…ì¥ëª…"] + "_" + df["ë©”ë‰´ëª…"]

    return df


def add_holiday_features(df):
    """
    ê³µíœ´ì¼ ê´€ë ¨ íŒŒìƒ ë³€ìˆ˜ ì¶”ê°€
    - ê³µíœ´ì¼_ì „ë‚ : ë‹¤ìŒ ë‚ ì´ ê³µíœ´ì¼ì´ë©´ 1, ì•„ë‹ˆë©´ 0
    - ê³µíœ´ì¼_ì§í›„: ì´ì „ ë‚ ì´ ê³µíœ´ì¼ì´ë©´ 1, ì•„ë‹ˆë©´ 0
    """
    df = df.copy()

    # ê·¸ë£¹ë³„ë¡œ ì •ë ¬í•˜ì—¬ ë‚ ì§œ ìˆœì„œ ë³´ì¥
    df = df.sort_values(["group_id", "ì˜ì—…ì¼ì"])

    # ê³µíœ´ì¼_ì „ë‚ : ë‹¤ìŒ ë‚ ì´ ê³µíœ´ì¼ì¸ì§€ í™•ì¸
    df["ê³µíœ´ì¼_ì „ë‚ "] = df.groupby("group_id")["ê³µíœ´ì¼ì—¬ë¶€"].shift(-1).fillna(0).astype(int)

    # ê³µíœ´ì¼_ì§í›„: ì´ì „ ë‚ ì´ ê³µíœ´ì¼ì¸ì§€ í™•ì¸
    df["ê³µíœ´ì¼_ì§í›„"] = df.groupby("group_id")["ê³µíœ´ì¼ì—¬ë¶€"].shift(1).fillna(0).astype(int)

    return df


def cap_sales_outliers_iqr(df):
    """
    IQR ê¸°ë²•ì„ ì´ìš©í•œ ë§¤ì¶œìˆ˜ëŸ‰ ì´ìƒì¹˜ ì²˜ë¦¬
    - ë§¤ì¶œìˆ˜ëŸ‰_capped: IQR ê¸°ì¤€ìœ¼ë¡œ ì´ìƒì¹˜ë¥¼ ìƒí•œ/í•˜í•œê°’ìœ¼ë¡œ ì œí•œ
    """
    df = df.copy()
    df["ë§¤ì¶œìˆ˜ëŸ‰_capped"] = df["ë§¤ì¶œìˆ˜ëŸ‰"].astype(float)

    # ê° ê·¸ë£¹ë³„ë¡œ ì´ìƒì¹˜ ì²˜ë¦¬
    for group_id in df["group_id"].unique():
        group_mask = df["group_id"] == group_id
        group_data = df[group_mask]

        # ë§¤ì¶œì´ 0ë³´ë‹¤ í° ê°’ë“¤ë§Œ ì‚¬ìš©í•˜ì—¬ IQR ê³„ì‚°
        positive_sales = group_data[group_data["ë§¤ì¶œìˆ˜ëŸ‰"] > 0]["ë§¤ì¶œìˆ˜ëŸ‰"]

        if len(positive_sales) == 0:
            # ì–‘ìˆ˜ ë§¤ì¶œì´ ì—†ëŠ” ê²½ìš° ì›ë˜ê°’ ê·¸ëŒ€ë¡œ ìœ ì§€
            continue

        # Q1, Q3 ê³„ì‚°
        Q1 = positive_sales.quantile(0.25)
        Q3 = positive_sales.quantile(0.75)
        IQR = Q3 - Q1

        # ìƒí•œ, í•˜í•œ ê²½ê³„ê°’ ì„¤ì •
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # ì´ìƒì¹˜ ì²˜ë¦¬: ê²½ê³„ê°’ìœ¼ë¡œ ì œí•œ
        df.loc[group_mask, "ë§¤ì¶œìˆ˜ëŸ‰_capped"] = df.loc[group_mask, "ë§¤ì¶œìˆ˜ëŸ‰"].clip(
            lower=lower_bound, upper=upper_bound
        )

    return df


def preprocess_sales_data_complete(df, peak_season_map=None):
    """
    ì™„ì „í•œ ë°ì´í„° ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸
    ê¸°ì¡´ ì „ì²˜ë¦¬ + ê³µíœ´ì¼ íŒŒìƒë³€ìˆ˜ + ì´ìƒì¹˜ ì²˜ë¦¬
    """
    # ê¸°ì¡´ ì „ì²˜ë¦¬ ë‹¨ê³„ë“¤
    df = preprocess_sales_data_first(df)
    df = preprocess_sales_data_second(df, peak_season_map)

    # ìƒˆë¡œìš´ ê¸°ëŠ¥ë“¤ ì¶”ê°€
    df = add_holiday_features(df)
    df = cap_sales_outliers_iqr(df)

    return df
````

````python
# ì‹œê³„ì—´ ë°ì´í„° ìƒì„± í•¨ìˆ˜
def create_time_series_features(df):
    """ì‹œê³„ì—´ í”¼ì²˜ ìƒì„± (ì´ë™í‰ê·  ì œì™¸)"""
    df = df.copy()

    # ë‚ ì§œ ê´€ë ¨ í”¼ì²˜
    df['ì¼'] = df['ì˜ì—…ì¼ì'].dt.day
    df['ë…„'] = df['ì˜ì—…ì¼ì'].dt.year
    df['ì›”_sin'] = np.sin(2 * np.pi * df['ì›”'] / 12)
    df['ì›”_cos'] = np.cos(2 * np.pi * df['ì›”'] / 12)
    df['ìš”ì¼_sin'] = np.sin(2 * np.pi * df['ìš”ì¼'] / 7)
    df['ìš”ì¼_cos'] = np.cos(2 * np.pi * df['ìš”ì¼'] / 7)

    # ë¼ê·¸ í”¼ì²˜ (1ì¼ì „, 7ì¼ì „, 14ì¼ì „, 21ì¼ì „)
    for group_id, group_data in df.groupby('group_id'):
        group_data = group_data.sort_values('ì˜ì—…ì¼ì')
        for lag in [1, 7, 14, 21]:
            df.loc[df['group_id'] == group_id, f'lag_{lag}'] = group_data['ë§¤ì¶œìˆ˜ëŸ‰'].shift(lag)

    # ê²°ì¸¡ê°’ ì²˜ë¦¬ (ë¼ê·¸ í”¼ì²˜)
    lag_cols = [f'lag_{lag}' for lag in [1, 7, 14, 21]]
    df[lag_cols] = df[lag_cols].fillna(0)

    return df

def create_sequences(df, lookback=28, predict=7):
    """ì‹œê³„ì—´ ì‹œí€€ìŠ¤ ìƒì„±"""
    sequences = []
    targets = []
    group_ids = []

    feature_cols = ['ë§¤ì¶œìˆ˜ëŸ‰', 'ìš”ì¼', 'ì›”', 'ì£¼ë§ì—¬ë¶€', 'ê³µíœ´ì¼ì—¬ë¶€',
                    'ì›”_sin', 'ì›”_cos', 'ìš”ì¼_sin', 'ìš”ì¼_cos',
                    'lag_1', 'lag_7', 'lag_14', 'lag_21']

    # ì„±ìˆ˜ê¸°ì—¬ë¶€ ì›í•« ì¸ì½”ë”©
    df_encoded = pd.get_dummies(df, columns=['ì„±ìˆ˜ê¸°ì—¬ë¶€'], prefix='season')
    season_cols = [col for col in df_encoded.columns if col.startswith('season_')]
    feature_cols.extend(season_cols)

    for group_id, group_data in df_encoded.groupby('group_id'):
        group_data = group_data.sort_values('ì˜ì—…ì¼ì').reset_index(drop=True)

        for i in range(len(group_data) - lookback - predict + 1):
            # ì…ë ¥ ì‹œí€€ìŠ¤ (lookbackì¼)
            seq = group_data.iloc[i:i+lookback][feature_cols].values
            # íƒ€ê²Ÿ (predictì¼ê°„ì˜ ë§¤ì¶œìˆ˜ëŸ‰)
            target = group_data.iloc[i+lookback:i+lookback+predict]['ë§¤ì¶œìˆ˜ëŸ‰'].values

            sequences.append(seq)
            targets.append(target)
            group_ids.append(group_id)

    return np.array(sequences), np.array(targets), group_ids, feature_cols
````

````python
# Multi-variate Time Series Transformer ëª¨ë¸
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_seq_length=1000):
        super(PositionalEncoding, self).__init__()

        pe = torch.zeros(max_seq_length, d_model)
        position = torch.arange(0, max_seq_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() *
                            (-torch.log(torch.tensor(10000.0)) / d_model))

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0).transpose(0, 1)

        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:x.size(0), :]

class TimeSeriesTransformer(nn.Module):
    def __init__(self, input_dim, d_model=128, n_heads=8, n_layers=4, d_ff=512,
                dropout=0.1, lookback=28, predict=7):
        super(TimeSeriesTransformer, self).__init__()

        self.input_dim = input_dim
        self.d_model = d_model
        self.lookback = lookback
        self.predict = predict

        # ì…ë ¥ íˆ¬ì˜
        self.input_projection = nn.Linear(input_dim, d_model)

        # ìœ„ì¹˜ ì¸ì½”ë”©
        self.positional_encoding = PositionalEncoding(d_model, max_seq_length=lookback)

        # Transformer Encoder
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=n_heads,
            dim_feedforward=d_ff,
            dropout=dropout,
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)

        # ì¶œë ¥ ë ˆì´ì–´
        self.output_projection = nn.Sequential(
            nn.Linear(d_model, d_ff),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(d_ff, predict)
        )

        # ë“œë¡­ì•„ì›ƒ
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        # x shape: (batch_size, seq_len, input_dim)
        batch_size, seq_len, _ = x.shape

        # ì…ë ¥ íˆ¬ì˜
        x = self.input_projection(x)  # (batch_size, seq_len, d_model)

        # ìœ„ì¹˜ ì¸ì½”ë”©
        x = x.transpose(0, 1)  # (seq_len, batch_size, d_model)
        x = self.positional_encoding(x)
        x = x.transpose(0, 1)  # (batch_size, seq_len, d_model)

        # ë“œë¡­ì•„ì›ƒ
        x = self.dropout(x)

        # Transformer Encoder
        x = self.transformer_encoder(x)  # (batch_size, seq_len, d_model)

        # ê¸€ë¡œë²Œ í’€ë§ (í‰ê· )
        x = x.mean(dim=1)  # (batch_size, d_model)

        # ì¶œë ¥ íˆ¬ì˜
        output = self.output_projection(x)  # (batch_size, predict)

        return output
````

````python
# í‰ê°€ ì§€í‘œ í•¨ìˆ˜ë“¤
def calculate_smape(y_true, y_pred):
    """SMAPE (Symmetric Mean Absolute Percentage Error) ê³„ì‚°"""
    numerator = np.abs(y_pred - y_true)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2

    # ë¶„ëª¨ê°€ 0ì¸ ê²½ìš° ì²˜ë¦¬ (ì‹¤ì œê°’ê³¼ ì˜ˆì¸¡ê°’ì´ ëª¨ë‘ 0)
    mask = denominator != 0
    smape_values = np.zeros_like(numerator)
    smape_values[mask] = numerator[mask] / denominator[mask]

    return np.mean(smape_values) * 100

def calculate_mae(y_true, y_pred):
    """MAE (Mean Absolute Error) ê³„ì‚°"""
    return np.mean(np.abs(y_true - y_pred))

def calculate_rmse(y_true, y_pred):
    """RMSE (Root Mean Square Error) ê³„ì‚°"""
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def calculate_mape(y_true, y_pred):
    """MAPE (Mean Absolute Percentage Error) ê³„ì‚° (ì˜ë§¤ì¶œ ì œì™¸)"""
    mask = y_true != 0
    if mask.sum() == 0:
        return 0.0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def evaluate_model(y_true, y_pred):
    """ëª¨ë¸ í‰ê°€"""
    metrics = {
        'MAE': calculate_mae(y_true, y_pred),
        'RMSE': calculate_rmse(y_true, y_pred),
        'SMAPE': calculate_smape(y_true, y_pred),
        'MAPE': calculate_mape(y_true, y_pred)
    }
    return metrics
````

````python
# ModelConfigë¥¼ ì‚¬ìš©í•œ ëª¨ë¸ ìƒì„± ë° í•™ìŠµ í•¨ìˆ˜
def create_model_from_config(config, input_dim, device):
    """Configë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ ìƒì„±"""
    model = TimeSeriesTransformer(
        input_dim=input_dim,
        d_model=config.d_model,
        n_heads=config.n_heads,
        n_layers=config.n_layers,
        d_ff=config.d_ff,
        dropout=config.dropout,
        lookback=config.lookback,
        predict=config.predict
    ).to(device)

    return model

def create_optimizer_from_config(model, config):
    """Configë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜µí‹°ë§ˆì´ì € ìƒì„±"""
    if config.optimizer_type == 'Adam':
        optimizer = torch.optim.Adam(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay
        )
    elif config.optimizer_type == 'AdamW':
        optimizer = torch.optim.AdamW(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay
        )
    elif config.optimizer_type == 'SGD':
        optimizer = torch.optim.SGD(
            model.parameters(),
            lr=config.learning_rate,
            weight_decay=config.weight_decay,
            momentum=0.9
        )
    else:
        raise ValueError(f"Unsupported optimizer: {config.optimizer_type}")

    return optimizer

def create_scheduler_from_config(optimizer, config):
    """Configë¥¼ ì‚¬ìš©í•˜ì—¬ ìŠ¤ì¼€ì¤„ëŸ¬ ìƒì„±"""
    if config.scheduler_type == 'ReduceLROnPlateau':
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer,
            mode='min',
            factor=config.scheduler_factor,
            patience=config.scheduler_patience,
            min_lr=config.min_lr
        )
    elif config.scheduler_type == 'StepLR':
        scheduler = torch.optim.lr_scheduler.StepLR(
            optimizer,
            step_size=30,
            gamma=0.1
        )
    elif config.scheduler_type == 'CosineAnnealingLR':
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
            optimizer,
            T_max=config.epochs
        )
    else:
        scheduler = None

    return scheduler

def create_criterion_from_config(config):
    """Configë¥¼ ì‚¬ìš©í•˜ì—¬ ì†ì‹¤í•¨ìˆ˜ ìƒì„±"""
    if config.criterion_type == 'MSELoss':
        criterion = nn.MSELoss()
    elif config.criterion_type == 'MAELoss':
        criterion = nn.L1Loss()
    elif config.criterion_type == 'HuberLoss':
        criterion = nn.HuberLoss(delta=1.0)
    elif config.criterion_type == 'SmoothL1Loss':
        criterion = nn.SmoothL1Loss()
    else:
        raise ValueError(f"Unsupported criterion: {config.criterion_type}")

    return criterion
````

````python
# ğŸ”§ ê°œì„ ëœ ModelConfig ê¸°ë°˜ í•™ìŠµ í•¨ìˆ˜ (ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ ì§€ì›)
def train_model_with_config(config, X_train, X_val, y_train, y_val, feature_cols, group_ids_train, group_ids_val, target_scalers, device):
    """
    ModelConfigë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì„ í•™ìŠµí•˜ëŠ” í•¨ìˆ˜ (ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ ì§€ì›)

    Args:
        config: ModelConfig ê°ì²´
        X_train, X_val, y_train, y_val: í•™ìŠµ/ê²€ì¦ ë°ì´í„°
        feature_cols: í”¼ì²˜ ì»¬ëŸ¼ ì´ë¦„ë“¤
        group_ids_train, group_ids_val: í•™ìŠµ/ê²€ì¦ ê·¸ë£¹ IDë“¤
        target_scalers: ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ ë”•ì…”ë„ˆë¦¬
        device: ë””ë°”ì´ìŠ¤ (CPU/GPU)

    Returns:
        model: í•™ìŠµëœ ëª¨ë¸
        history: í•™ìŠµ ê¸°ë¡ (train_losses, val_losses)
        metrics: í‰ê°€ ì§€í‘œ
    """

    print(f"=== ğŸ”§ ê°œì„ ëœ {config.__class__.__name__}ë¡œ ëª¨ë¸ í•™ìŠµ ì‹œì‘ ===")
    config.print_config()

    # 1. ë°ì´í„°ë¡œë” ìƒì„±
    train_dataset = TensorDataset(
        torch.FloatTensor(X_train).to(device),
        torch.FloatTensor(y_train).to(device)
    )
    val_dataset = TensorDataset(
        torch.FloatTensor(X_val).to(device),
        torch.FloatTensor(y_val).to(device)
    )

    train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=config.batch_size, shuffle=False)

    # 2. ëª¨ë¸, ì˜µí‹°ë§ˆì´ì €, ìŠ¤ì¼€ì¤„ëŸ¬, ì†ì‹¤í•¨ìˆ˜ ìƒì„±
    model = create_model_from_config(config, len(feature_cols), device)
    optimizer = create_optimizer_from_config(model, config)
    scheduler = create_scheduler_from_config(optimizer, config)
    criterion = create_criterion_from_config(config)

    print(f"ëª¨ë¸ íŒŒë¼ë¯¸í„° ìˆ˜: {sum(p.numel() for p in model.parameters()):,}")

    # 3. í•™ìŠµ ë£¨í”„
    train_losses = []
    val_losses = []
    best_val_loss = float('inf')
    patience_counter = 0

    for epoch in range(config.epochs):
        # í•™ìŠµ ëª¨ë“œ
        model.train()
        train_loss = 0.0

        for batch_x, batch_y in train_loader:
            optimizer.zero_grad()
            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=config.max_norm)
            optimizer.step()
            train_loss += loss.item()

        # ê²€ì¦ ëª¨ë“œ
        model.eval()
        val_loss = 0.0

        with torch.no_grad():
            for batch_x, batch_y in val_loader:
                outputs = model(batch_x)
                loss = criterion(outputs, batch_y)
                val_loss += loss.item()

        # í‰ê·  ì†ì‹¤ ê³„ì‚°
        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)

        # ìŠ¤ì¼€ì¤„ëŸ¬ ìŠ¤í…
        if scheduler and config.scheduler_type == 'ReduceLROnPlateau':
            scheduler.step(avg_val_loss)
        elif scheduler:
            scheduler.step()

        # Early Stopping
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            patience_counter = 0
            # ìµœê³  ëª¨ë¸ ì €ì¥
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1

        # ì§„í–‰ìƒí™© ì¶œë ¥ (ë§¤ 1 ì—í¬í¬ë§ˆë‹¤)
        if (epoch + 1) % 1 == 0:
            current_lr = optimizer.param_groups[0]['lr']
            print(f'Epoch [{epoch+1:3d}/{config.epochs}] | '
                    f'Train: {avg_train_loss:.6f} | Val: {avg_val_loss:.6f} | '
                    f'Best: {best_val_loss:.6f} | LR: {current_lr:.2e} | '
                    f'Patience: {patience_counter}/{config.early_stopping_patience}')

        # Early Stopping ì²´í¬
        if patience_counter >= config.early_stopping_patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

    # ìµœê³  ëª¨ë¸ ë³µì›
    model.load_state_dict(best_model_state)

    # 4. ğŸ”§ ê°œì„ ëœ ìµœì¢… í‰ê°€ (ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ ì‚¬ìš©)
    model.eval()
    val_predictions = []
    val_targets = []

    with torch.no_grad():
        for batch_x, batch_y in val_loader:
            outputs = model(batch_x)
            val_predictions.append(outputs.cpu().numpy())
            val_targets.append(batch_y.cpu().numpy())

    val_pred = np.concatenate(val_predictions, axis=0)
    val_true = np.concatenate(val_targets, axis=0)

    # ğŸ¯ í•µì‹¬ ìˆ˜ì •: ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ë¥¼ ì‚¬ìš©í•œ ì—­ë³€í™˜
    val_pred_orig = np.zeros_like(val_pred)
    val_true_orig = np.zeros_like(val_true)

    for i, group_id in enumerate(group_ids_val):
        if group_id in target_scalers:
            # ğŸ¯ ìˆ˜ì •ëœ ë¶€ë¶„: target_scalers ì‚¬ìš©
            scaler = target_scalers[group_id]
            
            # ì˜ˆì¸¡ê°’ ì—­ë³€í™˜
            pred_data = val_pred[i].reshape(-1, 1)
            val_pred_orig[i] = scaler.inverse_transform(pred_data).flatten()
            
            # ì‹¤ì œê°’ ì—­ë³€í™˜
            true_data = val_true[i].reshape(-1, 1)
            val_true_orig[i] = scaler.inverse_transform(true_data).flatten()
        else:
            # ìŠ¤ì¼€ì¼ëŸ¬ê°€ ì—†ëŠ” ê²½ìš° ì›ê°’ ì‚¬ìš©
            val_pred_orig[i] = val_pred[i]
            val_true_orig[i] = val_true[i]

    # í‰ê°€ ì§€í‘œ ê³„ì‚°
    metrics = evaluate_model(val_true_orig.flatten(), val_pred_orig.flatten())

    print("=== ìµœì¢… ì„±ëŠ¥ ===")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")
    
    # ğŸ” ì—­ë³€í™˜ ê²°ê³¼ ê²€ì¦
    print(f"\n=== ğŸ” ì—­ë³€í™˜ ê²°ê³¼ ê²€ì¦ ===")
    print(f"ì˜ˆì¸¡ê°’ ë²”ìœ„: {val_pred_orig.min():.2f} ~ {val_pred_orig.max():.2f}")
    print(f"ì‹¤ì œê°’ ë²”ìœ„: {val_true_orig.min():.2f} ~ {val_true_orig.max():.2f}")
    negative_preds = (val_pred_orig < 0).sum()
    total_preds = val_pred_orig.size
    print(f"ìŒìˆ˜ ì˜ˆì¸¡ê°’: {negative_preds}/{total_preds} ({negative_preds/total_preds*100:.2f}%)")

    history = {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'best_val_loss': best_val_loss
    }

    return model, history, metrics
````

````python
# ğŸ”§ ê°œì„ ëœ ëª¨ë¸ í‰ê°€ (ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ ì‚¬ìš©)
print("=== ğŸ”§ ê°œì„ ëœ ëª¨ë¸ í‰ê°€ ì‹œì‘ ===")

model.eval()
all_predictions = []
all_targets = []

with torch.no_grad():
    # í•™ìŠµ ë°ì´í„° ì˜ˆì¸¡
    for batch_x, batch_y in train_loader:
        outputs = model(batch_x)
        all_predictions.append(outputs.cpu().numpy())
        all_targets.append(batch_y.cpu().numpy())

    # ê²€ì¦ ë°ì´í„° ì˜ˆì¸¡
    val_predictions = []
    val_targets = []
    for batch_x, batch_y in val_loader:
        outputs = model(batch_x)
        val_predictions.append(outputs.cpu().numpy())
        val_targets.append(batch_y.cpu().numpy())

# ì˜ˆì¸¡ ê²°ê³¼ í•©ì¹˜ê¸°
train_pred = np.concatenate(all_predictions, axis=0)
train_true = np.concatenate(all_targets, axis=0)
val_pred = np.concatenate(val_predictions, axis=0)
val_true = np.concatenate(val_targets, axis=0)

print(f"ì˜ˆì¸¡ ì™„ë£Œ: í•™ìŠµ ë°ì´í„° {train_pred.shape}, ê²€ì¦ ë°ì´í„° {val_pred.shape}")

# ğŸ¯ í•µì‹¬ ìˆ˜ì •: ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ë¥¼ ì‚¬ìš©í•œ ì—­ë³€í™˜
print("ğŸ”§ ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ë¥¼ ì‚¬ìš©í•˜ì—¬ ì—­ë³€í™˜ ì¤‘...")

# í•™ìŠµ ë°ì´í„° ì—­ë³€í™˜
train_pred_orig = np.zeros_like(train_pred)
train_true_orig = np.zeros_like(train_true)

for i, group_id in enumerate(tqdm(group_ids_train, desc="í•™ìŠµ ë°ì´í„° ì—­ë³€í™˜")):
    if group_id in target_scalers:
        # ğŸ¯ ìˆ˜ì •ëœ ë¶€ë¶„: feature_scalers ëŒ€ì‹  target_scalers ì‚¬ìš©
        scaler = target_scalers[group_id]
        
        # ì˜ˆì¸¡ê°’ ì—­ë³€í™˜
        pred_data = train_pred[i].reshape(-1, 1)
        real_pred = scaler.inverse_transform(pred_data).flatten()
        train_pred_orig[i] = real_pred
        
        # ì‹¤ì œê°’ ì—­ë³€í™˜  
        true_data = train_true[i].reshape(-1, 1)
        real_true = scaler.inverse_transform(true_data).flatten()
        train_true_orig[i] = real_true
    else:
        # ìŠ¤ì¼€ì¼ëŸ¬ê°€ ì—†ëŠ” ê²½ìš° ì›ê°’ ì‚¬ìš©
        train_pred_orig[i] = train_pred[i]
        train_true_orig[i] = train_true[i]

# ê²€ì¦ ë°ì´í„° ì—­ë³€í™˜
val_pred_orig = np.zeros_like(val_pred)
val_true_orig = np.zeros_like(val_true)

for i, group_id in enumerate(tqdm(group_ids_val, desc="ê²€ì¦ ë°ì´í„° ì—­ë³€í™˜")):
    if group_id in target_scalers:
        # ğŸ¯ ìˆ˜ì •ëœ ë¶€ë¶„: feature_scalers ëŒ€ì‹  target_scalers ì‚¬ìš©
        scaler = target_scalers[group_id]
        
        # ì˜ˆì¸¡ê°’ ì—­ë³€í™˜
        pred_data = val_pred[i].reshape(-1, 1)
        real_pred = scaler.inverse_transform(pred_data).flatten()
        val_pred_orig[i] = real_pred
        
        # ì‹¤ì œê°’ ì—­ë³€í™˜
        true_data = val_true[i].reshape(-1, 1)
        real_true = scaler.inverse_transform(true_data).flatten()
        val_true_orig[i] = real_true
    else:
        # ìŠ¤ì¼€ì¼ëŸ¬ê°€ ì—†ëŠ” ê²½ìš° ì›ê°’ ì‚¬ìš©
        val_pred_orig[i] = val_pred[i] 
        val_true_orig[i] = val_true[i]

print("âœ… ê·¸ë£¹ë³„ íƒ€ê²Ÿ ìŠ¤ì¼€ì¼ëŸ¬ ì—­ë³€í™˜ ì™„ë£Œ")

# í‰ê°€ ì§€í‘œ ê³„ì‚°
print("\n=== í•™ìŠµ ë°ì´í„° ì„±ëŠ¥ ===")
train_metrics = evaluate_model(train_true_orig.flatten(), train_pred_orig.flatten())
for metric, value in train_metrics.items():
    print(f"{metric}: {value:.4f}")

print("\n=== ê²€ì¦ ë°ì´í„° ì„±ëŠ¥ ===")
val_metrics = evaluate_model(val_true_orig.flatten(), val_pred_orig.flatten())
for metric, value in val_metrics.items():
    print(f"{metric}: {value:.4f}")

# ğŸ” ì—­ë³€í™˜ ê²°ê³¼ ê²€ì¦
print(f"\n=== ğŸ” ì—­ë³€í™˜ ê²°ê³¼ ê²€ì¦ ===")
print(f"ê²€ì¦ ë°ì´í„° ì˜ˆì¸¡ê°’ ë²”ìœ„: {val_pred_orig.min():.2f} ~ {val_pred_orig.max():.2f}")
print(f"ê²€ì¦ ë°ì´í„° ì‹¤ì œê°’ ë²”ìœ„: {val_true_orig.min():.2f} ~ {val_true_orig.max():.2f}")
print(f"ì˜ˆì¸¡ê°’ í‰ê· : {val_pred_orig.mean():.2f}")
print(f"ì‹¤ì œê°’ í‰ê· : {val_true_orig.mean():.2f}")

# ìŒìˆ˜ ì˜ˆì¸¡ê°’ í™•ì¸
negative_predictions = (val_pred_orig < 0).sum()
total_predictions = val_pred_orig.size
print(f"ìŒìˆ˜ ì˜ˆì¸¡ê°’: {negative_predictions}/{total_predictions} ({negative_predictions/total_predictions*100:.2f}%)")

# ìš”ì¼ë³„ ì„±ëŠ¥ ë¶„ì„
print("\n=== ìš”ì¼ë³„ ì„±ëŠ¥ ë¶„ì„ (ê²€ì¦ ë°ì´í„°) ===")
for day in range(7):
    day_pred = val_pred_orig[:, day]
    day_true = val_true_orig[:, day]
    day_metrics = evaluate_model(day_true, day_pred)
    day_names = ['ì›”', 'í™”', 'ìˆ˜', 'ëª©', 'ê¸ˆ', 'í† ', 'ì¼']
    print(f"{day_names[day]}ìš”ì¼ - SMAPE: {day_metrics['SMAPE']:.2f}%, MAE: {day_metrics['MAE']:.2f}")

print("\nâœ… íƒ€ê²Ÿ ë³€ìˆ˜ ìŠ¤ì¼€ì¼ë§ ë²„ê·¸ ìˆ˜ì • ì™„ë£Œ!")
print("ğŸ¯ ì´ì œ ì˜ˆì¸¡ê°’ì´ ì˜¬ë°”ë¥¸ ë§¤ì¶œ ìˆ˜ëŸ‰ ë²”ìœ„ë¡œ ì¶œë ¥ë©ë‹ˆë‹¤.")
````
