# 1. LeetCode 225. Implement Stack using Queues

```python
class MyStack:

    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1

    def pop(self) -> int:
        return self.q1.popleft()

    def top(self) -> int:
        return self.q1[0]

    def empty(self) -> bool:
        return not self.q1
```

# 2. LeetCode 232. Implement Queue using Stacks

```python
class MyQueue:

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def transfer(self):
        while self.in_stack:
            self.out_stack.append(self.in_stack.pop())

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        if not self.out_stack:
            self.transfer()
        return self.out_stack.pop()

    def peek(self) -> int:
        if not self.out_stack:
            self.transfer()
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack
```

```python
from collections import deque

def is_in_language(s: str) -> bool:
    q = deque()
    found_dollar = False

    for ch in s:
        if ch == '$':
            found_dollar = True
            break
        q.append(ch)

    if not found_dollar:
        return False

    for ch in s[len(q)+1:]:
        if not q or q.pop() != ch:
            return False
```

```python
from collections import deque

class StackUsingQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())

        popped = self.q1.popleft()

        self.q1, self.q2 = self.q2, self.q1

        return popped

    def top(self):
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())

        top_item = self.q1.popleft()
        self.q2.append(top_item)

        self.q1, self.q2 = self.q2, self.q1

        return top_item

    def isEmpty(self):
        return len(self.q1) == 0
```

```python
class InefficientQueue:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, x):
        while self.stack1:
            self.stack2.append(self.stack1.pop())
        self.stack1.append(x)

        while self.stack2:
            self.stack1.append(self.stack2.pop())

    def dequeue(self):
        if not self.stack1:
            raise IndexError("Queue is empty")
        return self.stack1.pop()
```

```python
class ListDeque:
    def __init__(self):
        self._queue = deque()

    def isEmpty(self) -> bool:
        return len(self._queue) == 0

    def addFront(self, x):
        self._queue.appendleft(x)

    def addRear(self, x):
        self._queue.append(x)

    def deleteFront(self):
        if self.isEmpty():
            return None
        return self._queue.popleft()

    def deleteRear(self):
        if self.isEmpty():
            return None
        return self._queue.pop()

    def getFront(self):
        if self.isEmpty():
            return None
        return self._queue[0]

    def getRear(self):
        if self.isEmpty():
            return None
        return self._queue[-1]

    def dequeueAll(self):
        self._queue.clear()

    def printQueue(self):
        print("Deque from front:", end=' ')
        for item in self._queue:
            print(item, end=' ')
        print()
```
